```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.callbacks import TensorBoard
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv("Data_Integration_Reduced_2.csv")
data_y= data['Emotion_Real']
data_X = data.drop(['No.','Emotion_Real'], axis=1)
```

```python
X=data_X.values
X.shape
```

```python
# Number of folds for k-fold cross-validation
k = 5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
# Lists to store accuracy for each fold
clases = ['Neutral', 'Happy', 'Angry', 'Sad']
fold_accuracies = []
all_y_true = []
all_y_pred = []
```

```python
log_dir = "C:/Users/Poslu/Documents/Multimodal Emotion Recognition/Machine Learning Models (Own Dataset)/TensorBoard_Log" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
for fold, (train_index, val_index) in enumerate(skf.split(X, data_y), 1):
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = data_y.iloc[train_index], data_y.iloc[val_index]

    # Reshape the input data to match the input_shape in the model
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_val = np.reshape(X_val, (X_val.shape[0], X_val.shape[1], 1))

    # Define and compile the model
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='softplus', input_shape=(X_train.shape[1], 1)))
    model.add(Conv1D(filters=128, kernel_size=3, activation='softplus'))
    model.add(Conv1D(filters=256, kernel_size=2, activation='softplus'))
    model.add(Conv1D(filters=512, kernel_size=2, activation='softplus'))
    model.add(Conv1D(filters=1024, kernel_size=1, activation='softplus'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(512, activation='softplus'))
    model.add(Dense(256, activation='softplus'))
    model.add(Dense(128, activation='softplus'))
    model.add(Dense(64, activation='softplus'))
    model.add(Dense(4, activation='softmax'))

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Train the model with TensorBoard callback
    model.fit(X_train, y_train, epochs=500, batch_size=32, validation_data=(X_val, y_val),
              callbacks=[tensorboard_callback])

    # Evaluate the model on the validation set
    y_pred = np.argmax(model.predict(X_val), axis=1)
    accuracy = accuracy_score(y_val, y_pred)
    fold_accuracies.append(accuracy)
    all_y_true.extend(y_val)
    all_y_pred.extend(y_pred)

    # Generate classification report and confusion matrix for each fold
    print(f"\nFold {fold}")
    print("Classification Report:")
    print(classification_report(y_val, y_pred, target_names=clases))

    #Plot confusion matrix using seaborn
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=['Neutral', 'Happy', 'Angry', 'Sad'],
                yticklabels=['Neutral', 'Happy', 'Angry', 'Sad'])
    plt.title("Confusion Matrix - Fold {}".format(fold))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    
    cm_percentage = cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] # Calculate percentages
    plt.subplot(1, 2, 2)
    sns.heatmap(cm_percentage, annot=True, fmt=".2%", cmap="Blues",
                xticklabels=['Neutral', 'Happy', 'Angry', 'Sad'],
                yticklabels=['Neutral', 'Happy', 'Angry', 'Sad'])
    plt.title("Confusion Matrix - Fold {} (Percentage)".format(fold))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
print("\nOverall Classification Report:")
print(classification_report(all_y_true, all_y_pred))

# Plot confusion matrix using seaborn for all folds
overall_cm = confusion_matrix(all_y_true, all_y_pred)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.heatmap(overall_cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Neutral', 'Happy', 'Angry', 'Sad'],
            yticklabels=['Neutral', 'Happy', 'Angry', 'Sad'])
plt.title("Overall Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

overall_cm_percentage = overall_cm.astype('float') / overall_cm.sum(axis=1)[:, np.newaxis] # Calculate percentages
plt.subplot(1, 2, 2)
sns.heatmap(overall_cm_percentage, annot=True, fmt=".2%", cmap="Blues",
            xticklabels=['Neutral', 'Happy', 'Angry', 'Sad'],
            yticklabels=['Neutral', 'Happy', 'Angry', 'Sad'])
plt.title("Overall Confusion Matrix (Percentage)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.savefig('Overall_Confusion_Matrix.png', bbox_inches='tight', dpi=600)
```
